## 1768. Merge Strings Alternately


## This is my solution

```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        s1 = word1
        s2 = word2
        s3 = ""

        if(len(s1)>=len(s2)):
            s = len(s2) #small length
            l = len(s1) #large length
            m = s1[s:l:1]  #large part of the large length start from small length to large length
        else:
            s = len(s1)
            l = len(s2)
            m = s2[s:l:1]

        for i in range(0, s):
            s3 += s1[i] + s2[i]

        return s3+m
```

## This is my solution

```python
class Solution(object):
    def findTheDifference(self, s, t): 
        
        a = list(s)
        b = list(t)    # size -> t>s so b>a

        for i in a:
            if i in b:#if an element of 'a' is present in 'b' then remove it
                b.remove(i)

        c = b.pop() #only one element left so that is the answer

        return c
```

## 28. Find the Index of the First Occurrence in a String


## This is my solution

```python
class Solution(object):
    def strStr(self, haystack, needle):
        
        a = haystack
        b = needle
        c = -1   
        l = len(b) #size b < a
        for i in range(len(a)):
            if b[0] == a[i]: 
                if a[i:i+l] == b:  #a[0:0+3] as first example  
                    c = i
                    break         # need only the first occurrence  
        return c 
```

## This is another solution

```python
class Solution(object):
    def strStr(self, haystack, needle):
        # makes sure we don't iterate through a substring that is shorter than needle
        for i in range(len(haystack) - len(needle) + 1):
            # check if any substring of haystack with the same length as needle is equal to needle
            if haystack[i : i+len(needle)] == needle:
                # if yes, we return the first index of that substring
                return i
        # if we exit the loop, return -1        
        return -1
```

## 242. Valid Anagram


## This is my solution

```python
class Solution(object):
    def isAnagram(self, s, t):
        a = list(s)
        if len(a) == len(t): #if both str same size
            for i in range(len(t)):
                if t[i] in a: #if 2nd str one element present in 1st str then remove 1st str that element
                    a.remove(t[i])
            if len(a) == 0: #if 1st str is empty if removing 
                return True 
            else:
                return False
        else:   #if size is not same then they would not be same 
            return False
```

## This is another solution

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        sorted_s = sorted(s)
        sorted_t = sorted(t)
        return sorted_s == sorted_t
```

## This is another solution 2

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        count = [0] * 26
        
        # Count the frequency of characters in string s
        for x in s:
            count[ord(x) - ord('a')] += 1
        
        # Decrement the frequency of characters in string t
        for x in t:
            count[ord(x) - ord('a')] -= 1
        
        # Check if any character has non-zero frequency
        for val in count:
            if val != 0:
                return False
        
        return True
```

## 1822. Sign of the Product of an Array


## This is my solution

```python
class Solution(object):
    def arraySign(self, nums):
        a = 1       #because we multiple the list's numbers 
        for i in nums:
            a*=i    
        
        if a>=1:
            return 1
        elif a<=-1:
            return -1
        else:
            return 0
```

## This is another solution

```python
class Solution:
    def arraySign(self, nums: List[int]) -> int:
        ans = 1  # by default ans is 1
        for x in nums: 
            if x == 0: return 0 #if any value is 0 then definitely the product will be zero and the output will be 0
            if x < 0: ans *= -1  #if any value is less then 0 then we multiple -1 into the ans. Hint's is given below 
        return ans 
```

## 1502. Can Make Arithmetic Progression From Sequence


## This is my solution

```python
class Solution(object):
    def canMakeArithmeticProgression(self, arr): #let arr = [3,5,1]
        arr.sort(reverse=True)   #first it sorted then reversed 
        arr  # [5,3,1] 
        a = arr[0] - arr[1] # a = 5 - 3  -> 2
        for i in range(1,len(arr)-1): # 1 to last element - 1 
            if arr[i] - arr[i+1] !=a: #  if any two consecutive value is not match like a then it doesn't follow the rules 
                return False
        return True
```

## This is another solution

```python
class Solution:
    def canMakeArithmeticProgression(self, arr):
        length = len(arr)
        min_val = min(arr)
        max_val = max(arr)

        diff = (max_val - min_val) / (length - 1) #Calculate the common difference using this formula

        for i in range(length):
            expected = min_val + i * diff #Another formula to calculate the expected value, to get the list values. 
            if expected not in arr:
                return False

        return True
```

## Question 10
## 896. Monotonic Array


## This is my solution

```python
class Solution(object):
    def isMonotonic(self, nums):
        a = nums[:]   # saved a copy of the list
        b = sorted(nums) #saved the sorted list as variable b
        c = sorted(nums, reverse=True)# saved the sorted reversed list as variable c
    
        if b == a:  # is it sorted 
            return True
        elif c == a: # is it revered sorted 
            return True
        else:
            return False
```

## This is another solution

```python
class Solution(object):
    def isMonotonic(self, A):
        n = len(A)
        if n <= 2: return True #it has only one element so return true
				
        isGreat = False       
        isLess = False
        for i in range(1, n):  
            if A[i - 1] > A[i]: 
                isGreat = True
            if A[i - 1] < A[i]:
                isLess = True

            if isGreat and isLess: #if both are true that means they are not sorted so false
                return False

        return True #otherwise True
```

```python
class Solution:
    def toLowerCase(self, str):
        res = "" #taking an empty str
        for char in str:
            if ord(char) >= 65 and ord(char) <= 90: #ASCII codes for A-Z is 65-90 and those for a-z is that range plus 32.
                res += chr(ord(char)+32)
            else:
                res += char
        
        return res
```

## Question 14
## 682. Baseball Game



## This is my solution

```python
class Solution(object):
    def calPoints(self, operations):
        a = list()
        for i in operations:
            if i == 'C':
                a.pop()
            elif i == 'D':
                a.append(a[-1]*2)
            elif i == '+':
                a.append(a[-1]+a[-2])
            else:
                a.append(int(i))
        s = 0
        for i in a:
            s+=i
        return s
```

## This is another solution

```python
class Solution(object):
    def calPoints(self, ops):
        # Time: O(n)
        # Space: O(n)
        history = []
        for op in ops:
            if op == 'C':
                history.pop()
            elif op == 'D':
                history.append(history[-1] * 2)
            elif op == '+':
                history.append(history[-1] + history[-2])
            else:
                history.append(int(op))
        return sum(history)   #return total sum
```

## 657. Robot Return to Origin


## This is my solution 

```python
class Solution(object):
    def judgeCircle(self, moves):
        a = list(moves)
        for i in moves:
            if i == 'U':
                if 'D' in a:
                    a.remove('D')
            if i == 'D':
                if 'U' in a:
                    a.remove('U')
            if i == 'L':
                if 'R' in a:
                    a.remove('R')
            if i == 'R':
                if 'L' in a:
                    a.remove('L')

        if a:
            return False
        else:
            return True
#my logic is if you go up "U" you must need to down "D" or vise-versa. same for Right "R" and Left "L". 
#So to get True value if you have "UU" then you must need "DD". One 'D' for one 'U'
```
